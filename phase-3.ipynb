```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])
```

```python
print("==== Zomato sales analysis ==== ")
print("\n Sales data shape: ", sales_data.shape)
print("\n Data of first three rows: \n", sales_data[:3])
print("\n Pure data without id: \n", sales_data[:, 1:])
```

```python
print("Total sale in 2021: ", np.sum(sales_data[:,1], axis = 0)) #Sochke dekho
```

```python
total_sale = np.sum(sales_data[:, 1:], axis = 0)
print(total_sale)
```

```python
total_sale = np.sum(sales_data[:, 1:])
print(total_sale)
```

```python
# Pure year wise kitna sale hua sabka milake yani column wise so axis = 0
# Total sale of a restaurent ke case mein yani row wise axis = 1

min_sales = np.min(sales_data[:, 1:], axis = 1)
print(min_sales)
```

```python
# Tell me max sales per year 
max_sales = np.max(sales_data[:, 1:], axis = 0)
print(max_sales)
```

```python
# Tell me avg sale per restaurent
avg_sales = np.mean(sales_data[:, 1:], axis = 1)
print(avg_sales)
```

```python
# Tell me cumulative sales (yani har saal ek restaurent mein sale kaise ho rha)
cum_sale = np.cumsum(sales_data[:, 1:], axis = 1)
print(cum_sale)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(np.mean(cum_sale, axis=0))
plt.title("Average cummulative sale across all restaurent")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
